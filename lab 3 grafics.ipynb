```python
import pandas as pd 
from pathlib import Path

def get_data_df():
    df = pd.read_csv(r'D:\anaconda\lab 3\household_power_consumption.txt', sep = ';' )
    df.columns = ['Date', 'Time', 'GAP', 'GRP', 'Voltage', 'GI', 'SM1', 'SM2', 'SM3']    
    df = df.loc[(df['GAP'] != '?')]
    cols = ['GAP', 'GRP', 'Voltage', 'GI', 'SM1', 'SM2', 'SM3']
    df[cols] = df[cols].apply(pd.to_numeric)
    
    return df

df = get_data_df()

print(df)
```

```python
import pandas as pd 
import numpy as np

def get_data_arr():
    
    df = pd.read_csv(r'D:\anaconda\lab 3\household_power_consumption.txt', sep = ';' )
    df.columns = ['Date', 'Time', 'GAP', 'GRP', 'Voltage', 'GI', 'SM1', 'SM2', 'SM3']    
    df = df.loc[(df['GAP'] != '?')]
    cols = ['GAP', 'GRP', 'Voltage', 'GI', 'SM1', 'SM2', 'SM3']
    df[cols] = df[cols].apply(pd.to_numeric)
    array = df.values
    
    np.set_printoptions(edgeitems=5)
    
    return array

array = get_data_arr()

print (array)
```

```python
def filters_1_df(df): 
    df_filtered_1 = df.loc[(df['GAP'] > 5)]
    
    return df_filtered_1


filters_1_df(df)
```

```python
def filter_1_arr(array):
    
    array_filtered_1 = array[array[:,2] > 5] #GAP > 5
    
    return array_filtered_1 

print(filter_1_arr(array))
```

```python
import timeit
from matplotlib import pyplot as plt

x = []
y1= []
y2= []

for i in range (7):
    i = 10**(i)
    x.append(i)
    y1.append(timeit.timeit(f'filters_1_df(df[:{i}])' , number=100, globals=globals()))
    y2.append(timeit.timeit(f'filter_1_arr(array[:{i}])' , number=100, globals=globals()))
    
plt.plot(x,y1, label = 'pandas')
plt.plot(x,y2, label = 'numpy')
plt.xscale('log')
plt.legend()
```

```python
def split_df(df):
    
    df_filtered = df.loc[(df['Time'] >= '18:00:00') & (df['GAP'] >= 6) &
                                 (df['SM2'] > df['SM3']) & (df['SM2'] > df['SM1']), ]     

    r = len(df_filtered)
    
    df_1 = df_filtered[2:r // 2:3]
    df_2 = df_filtered[r // 2+1::4]
    
    return df_1, df_2

split_df(df)
```

```python
def split_arr(arr_filtered):
    
    arr_filtered = arr_filtered[(arr_filtered[:,1] > '18:00:00') & (arr_filtered[:,2] > 6) & 
                                (arr_filtered[:,7] > arr_filtered[:,6]) & (arr_filtered[:,7] > arr_filtered[:,8])]
    
    length = len(arr_filtered)
    
    arr_1 = arr_filtered[2:length//2:3] 
    arr_2 = arr_filtered[length//2+1::4] 

    return arr_1, arr_2

print (split_arr(arr_filtered))  
```

```python
x = []
y1= []
y2= []

for i in range (7):
    i = 10**(i)
    x.append(i)
    y1.append(timeit.timeit(f'split_df(df[:{i}])' , number=10, globals=globals()))
    y2.append(timeit.timeit(f'split_arr(arr_filtered[:{i}])' , number=10, globals=globals()))
    
plt.plot(x,y1, label = 'pandas')
plt.plot(x,y2, label = 'numpy')
plt.xscale('log')
plt.legend()
```
